## Generalization, Overfitting, and Underfitting
- In supervised learning, our goal is to build a model from the training data that can make accurate 
predictions on the test data (new unseen data). If we are able to achieve that, then we say our model is able to generalize 
well. However, that will always not be the case.

#### Overfitting:
        This is when our model performs well or achieves high accuracy on the training data used to build it but performs
    worse on new unseen data. This usually happens when the model is too complex for the data at hand 
    (that is, the data is too small for the model). Thus, the model memorizes the patterns in the training data but 
    lacks the ability to detect the patterns in the data it has not seen before. 
    
    Overfitting (high variance): means the model performed very well on the data used to build it( train data) but worst on the test data; remember our focus is more on the test. It means the model overfit the training data. it means it memorizes all the patterns including the noise (things not needed in the data) in the data and it was too weak to perform very well when we gave it a new data it hasn’t seen before. For instance, after memorizing a note for an exam and getting a question different from that in the note. Such model will give very high accuracy in the training so if you don’t check it on the test, you will think it is doing well. That is why it is important to always check the accuracy on the testing data too. Here, the model is too complex/powerful for the data.

#### Underfitting:
        This happens when our model is too simple such that it is not able to capture patterns even on the training data.
        Thus, it performs poorly on both the training data and the test data.
        
        underfitting (high bias): when the model is too weak to perform well on both the train and test. it is too weak to understand the patterns in the data. the model is too simple to understand the pattern in the training and as a result, it gives low accuracy in both training and testing. E.g student who did not read or memorized the note; even if the professor repeat questions in the note, the student will still fail. Here model is too simple compared to the data.
        
#### good compromise
 model performed well on the training and much more on the testing (higher): so it generalizes well. for instance our model is a good compromise. This is because it could get to this level even with one feature. because you are building the model for future, you have to focus on test data. Test data is a data the model hasn’t seen before.. You put more emphasis on the test data because it is an example of the future data that will be similar to the data used to build it and if the test data accuracy is higher then the model generalizes well
 
#### Regularization of Parameter: 
Parameter is where the name learning comes from. Models learn parameters of each variable in the data to predict the target. The algorithm will come up with a parameter for each of the variable (A1, B1,X1). if the parameters generated are too high, the model will become complex. Hence to reduce overfitting, we will reduce the values of the parameter the model suggested. This is known as regularization. For underfitting, we will increase the values of the parameter because the model is too simple.

#### resolving overfitting
- add more data if possible
- regularize model parameters
- stop model training early in terms of neural network

#### resolving underfitting
- increase the complexity of the model
- reduce regularization
- feature selection

### 8. MODEL REGULARIZATION
this is done in case the model didn’t perform well.

Multicollinearity: this occurs when two or more independent variables are highly correlated; hence variance becomes high. What we want is high correlation between the variables and target and not between variables and other variable. The implication is that the model will be confused on distinguishing between individual variables that is contributing to it in predicting the target. One way is to drop one of them. How to resolve
-	Remove some of the highly correlated independent variables.
-	Linearly combine the independent variables, such as adding them together
-	Regularization: it is going to look at the variables that are correlated and then tell the model not to pay attention to one of the variables that are correlated (tell it to ignore one of them). In that ways the model will improve. They include.

i.	Rigde regression: uses a penalty function call L2 to penalize the variable and reduce its β coefficient to nearly 0. It shrinks high-impact coefficient to nearly 0. Used to estimate regression model in case of multicollinearity. If we have 3 variables that are collinear with each other, laso will turn 2 of them to almost 0 <br>

ii.	Laso regression: it sets little impact to exactly 0. #### Model Evaluation
#### Function to measure Model Performances
- Model performance check
- We will be using metric functions defined in sklearn for RMSE, MAE, and R2.

- We will define functions to calculate adjusted R2.

- The adjuted R2 penalizes the regression model for adding unimportant features.
- We will create a function that will print out all the above metrics in one go.

```python
# function to compute adjusted R-squared
def adj_r2_score(predictors, targets, predictions): # The function takes the features, targetand the predctions
    r2 = r2_score(targets, predictions) #function that receives the target and prediction to produce the r2
    n = predictors.shape[0] # 0 means number of rows
    k = predictors.shape[1] # 1 (shape[1]) means number of variables
    return 1 - ((1 - r2) * (n - 1) / (n - k - 1))

# function to compute MAPE - Mean Absolute Percentage Error
def mape_score(targets, predictions):
    return np.mean(np.abs(targets - predictions) / targets) * 100


# function to compute different metrics to check performance of a regression model
def model_performance_regression(model, predictors, target):
    """
    Function to compute different metrics to check regression model performance

    model: regressor
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    r2 = r2_score(target, pred)  # to compute R-squared
    adjr2 = adj_r2_score(predictors, target, pred)  # to compute adjusted R-squared
    rmse = np.sqrt(mean_squared_error(target, pred))  # to compute RMSE
    mae = mean_absolute_error(target, pred)  # to compute MAE
    mape = mape_score(target, pred)  # to compute MAPE

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {
            "RMSE": rmse,
            "MAE": mae,
            "R-squared": r2,
            "Adj. R-squared": adjr2,
            "MAPE":mape,
        },
        index=[0],
    )

    return df_perf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
# Defining the function for creating boxplot and hisogram 
def histogram_boxplot(data, feature, figsize=(9, 5), kde=False, bins=None):
    """
    Boxplot and histogram combined
    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (12,7))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize)  # creating the 2 subplots
    
    sns.boxplot(data=data, x=feature, ax=ax_box2, showmeans=True, 
color="mediumturquoise")  # boxplot will be created and a star will indicate the 
mean value of the column
    
    if bins:
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, 
color="mediumpurple")
    else: 
      sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, 
color="mediumpurple")  # For histogram
    
    ax_hist2.axvline(data[feature].mean(), color="green", linestyle="--")  # Add 
mean to the histogram
    
    ax_hist2.axvline(data[feature].median(), color="black", linestyle="-")  # Add 
median to the histogram
    
# function to create labeled barplots
def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top
    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all 
levels)
    """
    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))
    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n].sort_values(),
    )
    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category
        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot
        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage
    plt.show()  # show the plot
```

#### Outlier Treatment

- Let's treat all the outliers by flooring and capping.

```python
# illustrate clipping values to minimum and maximum

l1 = [-200, 5, 0, 1000]
# 5 is representing upper whisker while 1000 is above it
# 0 is representing lower whisker while -200 is below it
# use np.clip function to clip the outliers to the minimum i.e. -200 to 0 and maximum i.e. 1000 to 5
np.clip(l1, 0, 5)
```

```python
def treat_outliers(df, col):
    """
    treats outliers in a varaible
    col: str, name of the numerical varaible
    df: data frame
    col: name of the column
    """
    Q1 = df[col].quantile(0.25)  # 25th quantile
    Q3 = df[col].quantile(0.75)  # 75th quantile
    IQR = Q3 - Q1  # Inter Quartile Range
    Lower_Whisker = Q1 - 1.5 * IQR # formular the boxplot used to determine the lower_whisker
    Upper_Whisker = Q3 + 1.5 * IQR
    df[col] = np.clip(
        df[col], Lower_Whisker, Upper_Whisker
    )  # all the values samller than Lower_Whisker will be assigned value of Lower_whisker
    # and all the values above upper_whisker will be assigned value of upper_Whisker
    return df


def treat_outliers_all(df, col_list):
    """
    treat outlier in all numerical varaibles
    col_list: list of numerical varaibles
    df: data frame
    """
    for c in col_list:
        df = treat_outliers(df, c)

    return df
```

It works best if there are no extreme values in the data and none of the actual values are 0.
    
* We will create a function which will print out all the above metrics in one go.

```python
# function to compute adjusted R-squared
def adj_r2_score(predictors, targets, predictions):
    r2 = r2_score(targets, predictions)
    n = predictors.shape[0]
    k = predictors.shape[1]
    return 1 - ((1 - r2) * (n - 1) / (n - k - 1))


# function to compute MAPE
def mape_score(targets, predictions):
    return np.mean(np.abs(targets - predictions) / targets) * 100


# function to compute different metrics to check performance of a regression model
def model_performance_regression(model, predictors, target):
    """
    Function to compute different metrics to check regression model performance

    model: regressor
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    r2 = r2_score(target, pred)  # to compute R-squared
    adjr2 = adj_r2_score(predictors, target, pred)  # to compute adjusted R-squared
    rmse = np.sqrt(mean_squared_error(target, pred))  # to compute RMSE
    mae = mean_absolute_error(target, pred)  # to compute MAE
    mape = mape_score(target, pred)  # to compute MAPE

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {
            "RMSE": rmse,
            "MAE": mae,
            "R-squared": r2,
            "Adj. R-squared": adjr2,
            "MAPE": mape
        },
        index=[0],
    )

    return df_perf
```

```python
# Checking model performance on train set
train_pf = model_performance_regression(lin_reg, x_train_sc, y_train)
train_pf
```
